```python
from PIL import Image
import numpy as np
import torch
from matplotlib import pyplot as plt


def get_image(image_path, target_width=1024, patch_size=256):
    """
    image_path から画像を読み込み、
    - 横幅を target_width にリサイズ
    - 高さを patch_size(=256) の倍数になるように下側にゼロパディング
    を行ってから torch.Tensor(C, H, W) を返す。
    """
    # 画像読み込み (RGB)
    img = Image.open(image_path).convert("RGB")

    # 元のサイズ
    orig_w, orig_h = img.size  # (W, H)

    # 横幅を target_width に変更し、高さは縦横比を維持してスケーリング
    new_w = target_width
    new_h = int(orig_h * new_w / orig_w)
    img = img.resize((new_w, new_h), Image.BILINEAR)

    # 高さを patch_size の倍数にパディング
    padded_h = ((new_h + patch_size - 1) // patch_size) * patch_size  # 切り上げ
    pad_bottom = padded_h - new_h

    if pad_bottom > 0:
        # (W, H+C) へのパディング用キャンバス（黒で埋める）
        padded = Image.new("RGB", (new_w, padded_h), (0, 0, 0))
        padded.paste(img, (0, 0))
        img = padded

    # Tensor 変換: (H, W, C) -> (C, H, W), float32
    tensor = torch.from_numpy(np.array(img)).permute(2, 0, 1).float()
    return tensor, orig_w, orig_h

# 使用例
image_path = get_image_path(target_images_path, images_id_list[i])
img, orig_w, orig_h = get_image(image_path, target_width=1024, patch_size=256)
plt.imshow(img.permute(1, 2, 0).byte())
plt.show()
img.shape
```

```python

def draw_rects_from_df(df, canvas_width, orig_w, orig_h):
    """
    df から矩形を読み取り、(H, W) の torch.float32 テンソルを返す。
    矩形領域は 1、それ以外は 0。
    """
    x_col="X"
    y_col="Y"
    w_col="Width"
    h_col="Height"

    org_height = orig_h
    org_width = orig_w
    canvas_height = int(org_height * canvas_width / org_width)
    canvas = torch.zeros((canvas_height, canvas_width), dtype=torch.float32)

    rate = canvas_width / org_width

    for _, row in df.iterrows():
        x = int(row[x_col] * rate)
        y = int(row[y_col] * rate)
        w = int(row[w_col] * rate)
        h = int(row[h_col] * rate)

        # 範囲チェック（キャンバス外にはみ出る場合の対策）
        x1 = max(x, 0)
        y1 = max(y, 0)
        x2 = min(x + w, canvas_width)
        y2 = min(y + h, canvas_height)

        if x1 < x2 and y1 < y2:
            canvas[y1:y2, x1:x2] = 1.0

    return canvas

# 例: 画像サイズが 1024x1024 の場合
canvas_width = 1024
canvas = draw_rects_from_df(annotation_data_df, canvas_width, orig_w, orig_h)
plt.imshow(canvas, cmap='gray')
plt.show()
```

```python

def draw_glue_from_df(df, canvas_width, orig_w, orig_h):
    """
    df から矩形を読み取り、(H, W) の torch.float32 テンソルを返す。
    矩形領域は 1、それ以外は 0。
    """
    x_col="X"
    y_col="Y"
    w_col="Width"
    h_col="Height"

    org_height = orig_h
    org_width = orig_w
    canvas_height = int(org_height * canvas_width / org_width)
    canvas = torch.zeros((canvas_height, canvas_width), dtype=torch.float32)

    rate = canvas_width / org_width

    for _, row in df.iterrows():
        x = int(row[x_col] * rate)
        y = int(row[y_col] * rate)
        w = int(row[w_col] * rate)
        h = int(row[h_col] * rate)
        min_l = int(min(w, h)*0.2)

        # 範囲チェック（キャンバス外にはみ出る場合の対策）
        x1 = max(x, 0)
        y1 = max(y, 0) - min_l
        x2 = min(x + w , canvas_width)
        y2 = min(y + h + min_l, canvas_height)

        if x1 < x2 and y1 < y2:
            canvas[y1:y2, x1:x2] += 1.0
    canvas = (canvas == 2).to(torch.float32)  # or .to(tensor.dtype)

    return canvas

# 例: 画像サイズが 1024x1024 の場合
canvas_width = 1024
canvas = draw_glue_from_df(annotation_data_df, canvas_width, orig_w, orig_h)
plt.imshow(canvas, cmap='gray')
plt.show()
```

```python
# ...existing code...
from torch.utils.data import Dataset, DataLoader

class KuzushijiCanvasDataset(Dataset):
    def __init__(self, root_dir, canvas_width=1024, transform=None, test_mode= False, test_docs=[]):
        """
        root_dir: char_sep_datas のパス
        canvas_width: キャンバス幅（高さは画像縦横比から計算）
        transform: 画像に対する torchvision.transforms など（任意）
        """
        self.root_dir = root_dir
        self.canvas_width = canvas_width
        self.transform = transform
        test_docs = set(test_docs)

        # ファイル一覧を作る
        # 形式: [(image_path, df_for_that_folder), ...] とする
        self.samples = []

        folders = get_folders(root_dir)
        for folder in folders:
            if test_mode:
                if folder not in test_docs:
                    continue
            else:
                if folder in test_docs:
                    continue
            target_doc, target_images_path = get_images_path(root_dir, folder)
            df = get_doc_df(target_doc)
            image_name_list = get_images_id(target_images_path)
            for image_name in image_name_list:
                image_path = get_image_path(target_images_path, image_name)
                image_id = get_image_id(image_name)  # 拡張子なし
                # DataFrame 側の Image 列と合わせる (必要に応じて調整)
                # 例: df['Image'] が拡張子なしならそのまま、拡張子付きなら ".png" を足すなど
                self.samples.append({
                    "image_path": image_path,
                    "image_id": image_id,
                    "df": df
                })

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        sample = self.samples[idx]
        image_path = sample["image_path"]
        image_id = sample["image_id"]
        df = sample["df"]

        # 画像ロード
        img, orig_w, orig_h = get_image(image_path, target_width=self.canvas_width,patch_size=256)  # (C, H, W) tensor

        img = img / 255.0  # 0-1 正規化
        # 必要なら変換
        if self.transform is not None:
            img = self.transform(img)

        # アノテーション抽出
        ann_df = get_annotation_data(df, image_id)

        # キャンバス生成
        canvas_main = draw_rects_from_df(ann_df, self.canvas_width, orig_w, orig_h)
        canvas_glue = draw_glue_from_df(ann_df, self.canvas_width, orig_w, orig_h)
        canvas = torch.stack([canvas_main, canvas_glue], dim=0)  # (2, Hc, Wc)

        return img, canvas

test_docs = [
    '200021637',
    '100249371',
    '100249537',
    '200005598',
    '200014740',
    '200020019',
    '200021712',
    '200021869'
]

# データセットとデータローダの作成例
train_dataset = KuzushijiCanvasDataset(
    root_dir=target_dir,
    canvas_width=2048,
    transform=None,  # ここに正規化などを入れてもよい
    test_mode=False,
    test_docs=test_docs
)
test_dataset = KuzushijiCanvasDataset(
    root_dir=target_dir,
    canvas_width=2048,
    transform=None,  # ここに正規化などを入れてもよい
    test_mode=True,
    test_docs=test_docs
)

train_dataloader = DataLoader(
    train_dataset,
    batch_size=1,
    shuffle=True,
    num_workers=0  # Windows なら最初は 0 推奨
)
test_dataloader = DataLoader(
    test_dataset,
    batch_size=1,
    shuffle=False,
    num_workers=0  # Windows なら最初は 0 推奨
)

# # 動作確認
# for imgs, canvases in dataloader:
#     print(imgs.shape)     # (B, C, H, W)
#     print(canvases.shape) # (B, 2, Hc, Wc)
    # break
# ...existing code...
```

```python
import torch.nn.functional as F

def patchify_image(img, patch_size=256):
    """
    img: (C, H, W) tensor
    return: patches: (N_patches, C, patch_size, patch_size)
    """
    C, H, W = img.shape
    assert H % patch_size == 0 and W % patch_size == 0, "H, W は patch_size の倍数を想定"

    # (1, C, H, W) -> unfold
    img = img.unsqueeze(0)
    patches = img.unfold(2, patch_size, patch_size).unfold(3, patch_size, patch_size)
    # shape: (1, C, H//p, W//p, p, p)
    patches = patches.contiguous().view(
        -1, C, patch_size, patch_size
    )  # (N_patches, C, p, p)
    return patches

def patchify_canvas(canvas, patch_size=256):
    """
    canvas: (K, Hc, Wc)  (K=2 チャネル)
    まず Hc, Wc を画像サイズ (H, W) に合わせておくことを推奨。
    ここでは単純に patch_size の倍数と仮定。
    return: (N_patches, K, patch_size, patch_size)
    """
    K, Hc, Wc = canvas.shape
    assert Hc % patch_size == 0 and Wc % patch_size == 0, "Hc, Wc は patch_size の倍数を想定"

    canvas = canvas.unsqueeze(0)  # (1, K, Hc, Wc)
    patches = canvas.unfold(2, patch_size, patch_size).unfold(3, patch_size, patch_size)
    patches = patches.contiguous().view(
        -1, K, patch_size, patch_size
    )  # (N_patches, K, p, p)
    return patches
```

# Transformer

```python
# ...existing code...
import torch.nn as nn

class PatchTransformer(nn.Module):
    def __init__(self, img_channels=3, patch_size=256, d_model=256,
                 nhead=8, num_layers=4, num_classes=2):
        super().__init__()
        self.patch_size = patch_size
        self.img_channels = img_channels

        patch_dim = img_channels * patch_size * patch_size  # 3*256*256
        self.patch_embed = nn.Linear(patch_dim, d_model)

        # 位置埋め込み（最大パッチ数を仮に 256 としておく）
        max_patches = 16384
        self.pos_embed = nn.Parameter(torch.zeros(1, max_patches, d_model))

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model, nhead=nhead, batch_first=True
        )
        self.encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

        # 各パッチからクラスを予測（例: 2クラス）
        self.cls_head = nn.Linear(d_model, num_classes)

    def forward(self, img):
        """
        img: (B, 3, H, W)
        return: logits: (B, N_patches, num_classes)
        """
        B, C, H, W = img.shape
        p = self.patch_size

        assert H % p == 0 and W % p == 0
        n_h = H // p
        n_w = W // p
        n_patches = n_h * n_w
        assert n_patches <= self.pos_embed.size(1), "max_patches を増やす必要あり"

        # パッチ分割: (B, C, H, W) -> (B, n_h*n_w, C*p*p)
        x = img.unfold(2, p, p).unfold(3, p, p)  # (B, C, n_h, n_w, p, p)
        x = x.contiguous().view(B, C, n_h * n_w, p * p)  # (B, C, N, p*p)
        x = x.permute(0, 2, 1, 3).contiguous().view(B, n_patches, -1)  # (B, N, C*p*p)
        # 埋め込み
        x = self.patch_embed(x)  # (B, N, d_model)

        # 位置埋め込み加算
        x = x + self.pos_embed[:, :n_patches, :]

        # Transformer Encoder
        x = self.encoder(x)  # (B, N, d_model)

        # 各パッチのクラス予測
        logits = self.cls_head(x)  # (B, N, num_classes)
        return logits
```

```python

def make_patch_labels(canvas, patch_size=256, threshold=0.5):
    """
    canvas: (2, Hc, Wc)
    glue チャネル (index=1) を見て、パッチの平均値 > threshold なら 1, それ以外 0
    return: labels: (N_patches,) long tensor
    """
    K, Hc, Wc = canvas.shape
    assert K == 2
    glue = canvas[1:2]  # (1, Hc, Wc)

    # パッチ分割
    patches = patchify_canvas(glue, patch_size=patch_size)  # (N, 1, p, p)
    # 平均値
    mean_vals = patches.mean(dim=(1, 2, 3))  # (N,)
    labels = (mean_vals > threshold).long()
    return labels

def make_patch_labels_multi(canvas, patch_size=256, threshold=0.3):
    """
    canvas: (2, Hc, Wc)  [0: main, 1: glue]
    各パッチについて、main / glue の平均値 > threshold なら 1, それ以外 0
    return: labels: (N_patches, 2) float tensor in {0,1}
    """
    K, Hc, Wc = canvas.shape
    assert K == 2

    # パッチ分割: (2, Hc, Wc) -> (N, 2, p, p)
    patches = patchify_canvas(canvas, patch_size=patch_size)  # (N, 2, p, p)

    # チャネルごとに平均値: (N, 2)
    mean_vals = patches.mean(dim=(2, 3))

    # threshold を超えたら 1、それ以外 0
    labels = (mean_vals > threshold).float()  # (N, 2)

    return labels
```

## 学習

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

patch_size = 256
model = PatchTransformer(
    img_channels=3,
    patch_size=patch_size,
    d_model=256,
    nhead=8,
    num_layers=4,
    num_classes=2    # 2チャネル: [main, glue]
).to(device)

criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

history = {'train_loss':[], 'val_loss':[]}

ckpt_dir = 'checkpoints_PatchTransformer'
os.makedirs(ckpt_dir, exist_ok=True)

num_epochs = 100
for epoch in range(num_epochs):  
    model.train()
    running_loss = 0.0
    n_batches = 0
    
    for imgs, canvases in train_dataloader:
        imgs = imgs.to(device)         # (B, 3, H, W)
        canvases = canvases.to(device) # (B, 2, Hc, Wc)

        B, C, H, W = imgs.shape

        # キャンバスを画像サイズにリサイズ → ピクセルごとの 0/1 ラベル
        target = F.interpolate(
            canvases, size=(H, W), mode="nearest"
        )  # (B, 2, H, W)

        # 順伝播
        logits = model(imgs)  # (B, 2, H, W)

        loss = criterion(logits, target.float())
        history.append(loss.item())

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        n_batches += 1

        # break 
 
    train_epoch_loss = running_loss / n_batches
    history['train_loss'].append(train_epoch_loss)

    model.eval()
    val_running_loss = 0.0
    val_n_batches = 0
    with torch.no_grad():
        for imgs, canvases in test_dataloader:
            imgs = imgs.to(device)
            canvases = canvases.to(device)

            B, C, H, W = imgs.shape
            target = F.interpolate(
                canvases, size=(H,W), mode= 'nearest'
            )

            logits = model(imgs)
            val_loss = criterion(logits, target.float())

            val_running_loss += val_loss.item()
            val_n_batches += 1
    val_epoch_loss = val_running_loss / val_n_batches
    history['val_loss'].append(val_epoch_loss)

    print(
        f'Epoch [{epoch+1}/{num_epochs}]',
        f'Train Loss: {train_epoch_loss:.4f} Val Loss: {val_epoch_loss:.4f}')

    ckpt_path = os.path.join(ckpt_dir, f'epoch_{epoch+1}.pth')
    torch.save(
        {
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'history': history,
        },
        ckpt_path,
    )
    print(f'saved checkpoint: {ckpt_path}')

# ...existing code...
plt.plot(history['train_loss'], label='train')
plt.plot(history['val_loss'], label='val')
plt.xlabel("Epoch")
plt.ylabel("loss")
plt.title("PatchTransformerSeg Loss")
plt.legend()
plt.show()
```

# U-Net

```python

# filepath: c:\Users\kotat\MyPrograms\MyKuzushiji\Glue_Craft\test\initial_trial.ipynb
# ...existing code...
import torch.nn as nn

class DoubleConv(nn.Module):
    """(Conv => BN => ReLU) x2"""
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.block = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
        )

    def forward(self, x):
        return self.block(x)


class UNet(nn.Module):
    def __init__(self, in_channels=3, num_classes=2, base_ch=64):
        super().__init__()
        # Encoder
        self.enc1 = DoubleConv(in_channels, base_ch)
        self.pool1 = nn.MaxPool2d(2)
        self.enc2 = DoubleConv(base_ch, base_ch * 2)
        self.pool2 = nn.MaxPool2d(2)
        self.enc3 = DoubleConv(base_ch * 2, base_ch * 4)
        self.pool3 = nn.MaxPool2d(2)
        self.enc4 = DoubleConv(base_ch * 4, base_ch * 8)
        self.pool4 = nn.MaxPool2d(2)

        # Bottleneck
        self.bottleneck = DoubleConv(base_ch * 8, base_ch * 16)

        # Decoder
        self.up4 = nn.ConvTranspose2d(base_ch * 16, base_ch * 8, kernel_size=2, stride=2)
        self.dec4 = DoubleConv(base_ch * 16, base_ch * 8)
        self.up3 = nn.ConvTranspose2d(base_ch * 8, base_ch * 4, kernel_size=2, stride=2)
        self.dec3 = DoubleConv(base_ch * 8, base_ch * 4)
        self.up2 = nn.ConvTranspose2d(base_ch * 4, base_ch * 2, kernel_size=2, stride=2)
        self.dec2 = DoubleConv(base_ch * 4, base_ch * 2)
        self.up1 = nn.ConvTranspose2d(base_ch * 2, base_ch, kernel_size=2, stride=2)
        self.dec1 = DoubleConv(base_ch * 2, base_ch)

        # 出力: num_classes チャネル (2チャネル: main/glue)
        self.out_conv = nn.Conv2d(base_ch, num_classes, kernel_size=1)

    def forward(self, x):
        # Encoder
        e1 = self.enc1(x)         # (B, 64, H, W)
        p1 = self.pool1(e1)       # (B, 64, H/2, W/2)
        e2 = self.enc2(p1)        # (B, 128, H/2, W/2)
        p2 = self.pool2(e2)       # (B, 128, H/4, W/4)
        e3 = self.enc3(p2)        # (B, 256, H/4, W/4)
        p3 = self.pool3(e3)       # (B, 256, H/8, W/8)
        e4 = self.enc4(p3)        # (B, 512, H/8, W/8)
        p4 = self.pool4(e4)       # (B, 512, H/16, W/16)

        # Bottleneck
        b = self.bottleneck(p4)   # (B, 1024, H/16, W/16)

        # Decoder
        u4 = self.up4(b)          # (B, 512, H/8, W/8)
        u4 = torch.cat([u4, e4], dim=1)  # skip connection
        d4 = self.dec4(u4)

        u3 = self.up3(d4)         # (B, 256, H/4, W/4)
        u3 = torch.cat([u3, e3], dim=1)
        d3 = self.dec3(u3)

        u2 = self.up2(d3)         # (B, 128, H/2, W/2)
        u2 = torch.cat([u2, e2], dim=1)
        d2 = self.dec2(u2)

        u1 = self.up1(d2)         # (B, 64, H, W)
        u1 = torch.cat([u1, e1], dim=1)
        d1 = self.dec1(u1)

        logits = self.out_conv(d1)  # (B, num_classes, H, W)
        return logits
# ...existing code...
```

# TransformerSeg

```python
class PatchTransformerSeg(nn.Module):
    def __init__(self, img_channels=3, patch_size=256, d_model=256,
                 nhead=8, num_layers=4, num_classes=2):
        super().__init__()
        self.patch_size = patch_size
        self.img_channels = img_channels

        patch_dim = img_channels * patch_size * patch_size
        self.patch_embed = nn.Linear(patch_dim, d_model)

        # 位置埋め込み（最大パッチ数を仮に 256 としておく）
        max_patches = 256
        self.pos_embed = nn.Parameter(torch.zeros(1, max_patches, d_model))

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model, nhead=nhead, batch_first=True
        )
        self.encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

        # --- ここから UpConv デコーダ ---
        # パッチグリッド (n_h, n_w) から段階的に解像度を上げる
        # 出力解像度は最後に F.interpolate で (H, W) に揃えるので、
        # アップサンプリング回数はそこまで厳密でなくてよいです。
        self.up1 = nn.ConvTranspose2d(d_model, d_model // 2, kernel_size=2, stride=2)
        self.up2 = nn.ConvTranspose2d(d_model // 2, d_model // 4, kernel_size=2, stride=2)
        self.up3 = nn.ConvTranspose2d(d_model // 4, d_model // 8, kernel_size=2, stride=2)

        self.dec_conv = nn.Sequential(
            nn.Conv2d(d_model // 8, d_model // 8, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(d_model // 8, num_classes, kernel_size=1)
        )

    def forward(self, img):
        """
        img: (B, 3, H, W)
        return: logits: (B, num_classes, H, W)  # ここでは num_classes=2 (main, glue)
        """
        B, C, H, W = img.shape
        p = self.patch_size

        assert H % p == 0 and W % p == 0
        n_h = H // p
        n_w = W // p
        n_patches = n_h * n_w
        assert n_patches <= self.pos_embed.size(1), "max_patches を増やす必要あり"

        # (B, C, H, W) -> パッチ列 (B, N, C*p*p)
        x = img.unfold(2, p, p).unfold(3, p, p)  # (B, C, n_h, n_w, p, p)
        x = x.contiguous().view(B, C, n_h * n_w, p * p)  # (B, C, N, p*p)
        x = x.permute(0, 2, 1, 3).contiguous().view(B, n_patches, -1)  # (B, N, C*p*p)

        # 線形埋め込み
        x = self.patch_embed(x)  # (B, N, d_model)

        # 位置埋め込み
        x = x + self.pos_embed[:, :n_patches, :]

        # Transformer Encoder
        x = self.encoder(x)  # (B, N, d_model)

        # パッチグリッドに戻す -> (B, d_model, n_h, n_w)
        x = x.view(B, n_h, n_w, -1).permute(0, 3, 1, 2).contiguous()

        x = F.relu(self.up1(x))  # (B, d_model//2, n_h*2,   n_w*2)
        x = F.relu(self.up2(x))  # (B, d_model//4, n_h*4,   n_w*4)
        x = F.relu(self.up3(x))  # (B, d_model//8, n_h*8,   n_w*8)

        # 2チャネル (main, glue) ロジット
        logits_low = self.dec_conv(x)  # (B, 2, H_low, W_low)

        # 最終的に元画像サイズにリサイズ
        logits = F.interpolate(logits_low, size=(H, W), mode="bilinear", align_corners=False)

        return logits
```

# TransformerSep

```python
import torch.nn as nn

class PatchTransformerSep(nn.Module):
    def __init__(self, img_channels=3, big_patch_size=256, vit_patch_size=16,
                 d_model=256, nhead=8, num_layers=4, num_classes=2):
        super().__init__()
        self.big_patch_size = big_patch_size     # 256
        self.vit_patch_size = vit_patch_size     # 16
        self.img_channels = img_channels

        # ViT 内部でのパッチの次元 (3 * 16 * 16)
        inner_patch_dim = img_channels * vit_patch_size * vit_patch_size
        self.patch_embed = nn.Linear(inner_patch_dim, d_model)

        # 各 256×256 タイルの中のトークン数 = (256/16)^2 = 256
        max_tokens = (big_patch_size // vit_patch_size) ** 2  # 256
        self.pos_embed = nn.Parameter(torch.zeros(1, max_tokens, d_model))

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model, nhead=nhead, batch_first=True
        )
        self.encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

        # タイル内の 16×16 グリッドを num_classes チャネルに射影
        self.head = nn.Conv2d(d_model, num_classes, kernel_size=1)

    def forward(self, img):
        """
        img: (B, 3, H, W),  H,W は 256 の倍数を想定
        return: logits: (B, num_classes, H, W)
        """
        B, C, H, W = img.shape
        P = self.big_patch_size      # 256
        p = self.vit_patch_size      # 16

        assert H % P == 0 and W % P == 0, "H,W は big_patch_size(256) の倍数を想定"
        nH_big = H // P   # 大タイル数 (縦)
        nW_big = W // P   # 大タイル数 (横)
        n_big = nH_big * nW_big

        # --- 1. 画像を 256×256 タイルに分割 ---
        # img: (B, C, H, W)
        x = img.unfold(2, P, P).unfold(3, P, P)  # (B, C, nH_big, nW_big, P, P)
        x = x.permute(0, 2, 3, 1, 4, 5).contiguous()  # (B, nH_big, nW_big, C, P, P)
        # for i in range(x.shape[1]):
        #     for j in range(x.shape[2]):
        #         plt.imshow(x[0, i, j].permute(1, 2, 0).cpu())
        #         plt.title(f"Tile ({i}, {j})")
        #         plt.axis("off")
        #         plt.show()
        x = x.view(B * n_big, C, P, P)  # (B*n_big, C, 256, 256)  各タイルをバッチ扱い

        # --- 2. 各 256×256 タイルを 16×16 パッチに分割（ViT 方式） ---
        n_h_inner = P // p  # 16
        n_w_inner = P // p  # 16
        N_inner = n_h_inner * n_w_inner  # 256

        # x: (B*n_big, C, 256, 256)
        x = x.unfold(2, p, p).unfold(3, p, p)  # (B*n_big, C, n_h_inner, n_w_inner, p, p)
        x = x.contiguous().view(
            B * n_big, C, N_inner, p * p
        )  # (B*n_big, C, 256, 16*16)
        x = x.permute(0, 2, 1, 3).contiguous().view(
            B * n_big, N_inner, -1
        )  # (B*n_big, 256, C*p*p)

        # --- 3. ViT 埋め込み + Transformer ---
        x = self.patch_embed(x)  # (B*n_big, 256, d_model)
        x = x + self.pos_embed[:, :N_inner, :]  # 位置埋め込み
        x = self.encoder(x)      # (B*n_big, 256, d_model)

        # --- 4. タイル内 16×16 グリッドに戻して 256×256 にアップサンプル ---
        x = x.view(B * n_big, n_h_inner, n_w_inner, -1).permute(
            0, 3, 1, 2
        ).contiguous()   # (B*n_big, d_model, 16, 16)

        logits_small = self.head(x)  # (B*n_big, num_classes, 16, 16)

        # 各タイル内を 256×256 に補間
        logits_tile = F.interpolate(
            logits_small, size=(P, P), mode="bilinear", align_corners=False
        )  # (B*n_big, num_classes, 256, 256)

        # --- 5. タイルを元位置に敷き詰めて (B, num_classes, H, W) を作る ---
        logits_tile = logits_tile.view(
            B, nH_big, nW_big, -1, P, P
        )  # (B, nH_big, nW_big, num_classes, 256, 256)

        logits = logits_tile.permute(
            0, 3, 1, 4, 2, 5
        ).contiguous().view(
            B, -1, H, W
        )  # (B, num_classes, H, W)
        # for i in range(logits.shape[0]):
        #     for j in range(logits.shape[1]):
        #         plt.title(f'batch[{i}] class[{j}]')
        #         plt.imshow(logits[i][j].cpu().detach().numpy())
        #         plt.show()

        return logits
```

## 学習

```python

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

patch_size = 256
model = PatchTransformerSep(
    img_channels=3,
    big_patch_size=256,
    vit_patch_size=16,
    d_model=256,
    nhead=8,
    num_layers=4,
    num_classes=2
).to(device)

criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

history = {'train_loss':[], 'val_loss':[]}

ckpt_dir = 'checkpoints'
os.makedirs(ckpt_dir, exist_ok=True)

num_epochs = 100
for epoch in range(num_epochs):  
    model.train()
    running_loss = 0.0
    n_batches = 0
    
    for imgs, canvases in train_dataloader:
        imgs = imgs.to(device)         # (B, 3, H, W)
        canvases = canvases.to(device) # (B, 2, Hc, Wc)

        B, C, H, W = imgs.shape

        # キャンバスを画像サイズにリサイズ → ピクセルごとの 0/1 ラベル
        target = F.interpolate(
            canvases, size=(H, W), mode="nearest"
        )  # (B, 2, H, W)

        # 順伝播
        logits = model(imgs)  # (B, 2, H, W)

        loss = criterion(logits, target.float())

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        n_batches += 1

        # break  # 動作確認用; 全バッチ回したいなら消す

    train_epoch_loss = running_loss / n_batches
    history['train_loss'].append(train_epoch_loss)

    model.eval()
    val_running_loss = 0.0
    val_n_batches = 0
    with torch.no_grad():
        for imgs, canvases in test_dataloader:
            imgs = imgs.to(device)
            canvases = canvases.to(device)

            B, C, H, W = imgs.shape
            target = F.interpolate(
                canvases, size=(H,W), mode= 'nearest'
            )

            logits = model(imgs)
            val_loss = criterion(logits, target.float())

            val_running_loss += val_loss.item()
            val_n_batches += 1
    val_epoch_loss = val_running_loss / val_n_batches
    history['val_loss'].append(val_epoch_loss)

    print(
        f'Epoch [{epoch+1}/{num_epochs}]',
        f'Train Loss: {train_epoch_loss:.4f} Val Loss: {val_epoch_loss:.4f}')

    ckpt_path = os.path.join(ckpt_dir, f'patchtransformersep_epoch_{epoch+1}.pth')
    torch.save(
        {
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'history': history,
        },
        ckpt_path,
    )
    print(f'saved checkpoint: {ckpt_path}')

# ...existing code...
plt.plot(history['train_loss'], label='train')
plt.plot(history['val_loss'], label='val')
plt.xlabel("Epoch")
plt.ylabel("loss")
plt.title("PatchTransformerSeg Loss")
plt.legend()
plt.show()
```
